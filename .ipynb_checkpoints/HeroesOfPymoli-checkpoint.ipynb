## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate unique items
items = purchase_data['Item ID'].nunique()

items
```

```python
# Calculate number of purchases
purchases = len(purchase_data)

purchases
```

```python
# Calculate total revenue
revenue = purchase_data['Price'].sum()

revenue
```

```python
# Create a summary data frame to hold the results
summary_total_df = pd.DataFrame({'Number of Unique Items': [items], 
                                 'Average Price': avg_price,
                                 'Number of Purchases': purchases, 
                                 'Total Revenue': revenue})

summary_total_df
```

```python
# Format to go to two decimal places, include a dollar sign, and use comma notation for revenue
summary_total_df['Average Price'] = summary_total_df['Average Price'].map('${:.2f}'.format)
summary_total_df['Total Revenue'] = summary_total_df['Total Revenue'].map('${:,.2f}'.format)

summary_total_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins for age groups
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100 ]

# Create labels for these bins
age_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
# Slice data and place it into bins
pd.cut(purchase_data['Age'], age_bins, labels=age_labels)
```

```python
# Append age bins to dataframe "purchase_data"
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'], age_bins, labels=age_labels)

purchase_data.head()
```

```python
# Group by age ranges
age_group = purchase_data.groupby('Age Ranges')

# Count unique players per age ranges
age_s = age_group['SN'].nunique()

age_s
```

```python
# Convert previously created series to dataframe
demographics_df = pd.DataFrame(age_s)

demographics_df
```

```python
# Remove index name ('Age Ranges')
demographics_df.index.name=None

# Rename column "SN"
demographics_df = demographics_df.rename(columns={'SN': 'Total Count'})

demographics_df
```

```python
# Create new column "Percentage of Players" and do cleaner format (add % sign)
demographics_df['Percentage of Players'] = ((demographics_df['Total Count'] / 
                                             total_players) * 100).map('{:.2f}%'.format)
demographics_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create dataframe for purchase count by players (SN)
player_group = purchase_data.groupby('SN')

# Count purchases per person
player_s = player_group['Purchase ID'].count()

player_s
```

```python
# Convert previously created series to dataframe
top_spenders_df = pd.DataFrame(player_s)

# Rename column "Purchase ID"
top_spenders_df = top_spenders_df.rename(columns={'Purchase ID': 'Purchase Count'})

top_spenders_df.head()
```

```python
# How many players bought particular number of items
top_spenders_df['Purchase Count'].value_counts()
```

```python
# Create new column with avg. purchase price by person (SN)
top_spenders_df['Average Purchase Price'] = player_group['Price'].mean()

# Create new column with total purchase value by person (SN)
top_spenders_df['Total Purchase Value'] = player_group['Price'].sum()

# Display top spenders - sort total purchase value in descendig order
top_spenders_df = top_spenders_df.sort_values(by='Total Purchase Value', ascending=False)

top_spenders_df.head()
```

```python
# Format to go to two decimal places and include a dollar sign
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map(
    '${:.2f}'.format)   
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map(
    '${:,.2f}'.format)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
items_df = purchase_data[['Item ID', 'Item Name', 'Price']]

items_df.head()
```

```python
# Group by Item ID and Item Name
items_group = items_df.groupby(['Item ID', 'Item Name'])

# Create dataframe
items_gb_df = pd.DataFrame(items_group['Price'].value_counts())

items_gb_df.head()
```

```python
# Rename column 'Price'
items_gb_df = items_gb_df.rename(columns={'Price': 'Purchase Count'})

# Rename axis for index 'Price'
items_gb_df = items_gb_df.rename_axis(index={'Price': 'Item Price'})

items_gb_df.head()
```

```python
# Remove level 'Item Price' from index
items_gb_df = items_gb_df.reset_index(level='Item Price')

items_gb_df.head()
```

```python
# Reorder columns
items_gb_df = items_gb_df[['Purchase Count', 'Item Price']]

items_gb_df.head()
```

```python
# Calculate total purchase value per item and add it as new column
items_gb_df['Total Purchase Value'] = items_gb_df['Purchase Count'] * items_gb_df['Item Price']

items_gb_df.head()
```

```python
# Display most popular items - sort purchase count in descending order
popular_df = items_gb_df.sort_values('Purchase Count', ascending=False)

popular_df.head()
```

```python
# Format to go to two decimal places and include a dollar sign
popular_df['Item Price'] = popular_df['Item Price'].map('${:.2f}'.format)   
popular_df['Total Purchase Value'] = popular_df['Total Purchase Value'].map('${:,.2f}'.format)

popular_df.head()
```

```python
# Display most profitable items - sort total purchase value in descending order 
profitable_df = items_gb_df.sort_values('Total Purchase Value', ascending=False)

profitable_df.head()
```

## Observable Trends



There are several trends that we could observe based on the game's purchasing data:

* **Gender Purchasing Analysis** - Male players make majority of players, 84% (484 players). This also means that male players made most purchases, but female and other players have spent about 10% more in average per person than male players. 


* **Demographics and Age Purchasing Analysis** - Majority of players (~45%) belong into the <i>"20-24"</i> age group and total counts distribution per age ranges very closely follows normal distribution. Total purchase value follows very similar distribution because average purchase prices per age ranges are all around \$3.00. Average purchase price is slightly higher only for two age ranges: <i>"<10"</i> is 12% and <i>"35-39"</i> is 20%.


* **Top Spenders** - Majority of players bought 1 or 2 items only, 414 and 124 players respectively. 35 players bought 3 items. Only three players bought 5 or 4 items, 1 and 2 players respectively. As expected, these 3 players are among 5 top spenders.


* **Most Popular and Most Profitable Items** - Three items are on both lists, five most popular and five most profitable items. Item prices are in the range from \\$3.5 and \\$4.9, except for one item. This item is priced only \$1.02 and it made only most popular list. It was bought 8 times but due to low price the profit was low too.
